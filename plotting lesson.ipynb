```python
import os
```

```python
distance_file= os.path.join('distance_data_headers.csv')
import numpy
distance = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers = distance[0,:]
data = distance [1:,:]
data = data.astype(numpy.float)
print(data)
print (headers)
```

```python
import matplotlib.pyplot as plt
```

```python
sample = headers[1]
sample2 = headers[2]

plt.plot(data[:,1], label=sample)
plt.plot(data[:,2], label=sample2)
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend()

plt.savefig('two_samples.png',dpi=300)
```

```python
num_columns=len(headers)
print(num_columns)

for columns in range(1,num_columns):
    plt.figure()
    plt.plot(data[:, columns], label=headers[columns])
    plt.legend()
    
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.savefig(F'{headers[columns]}.png', dpi=300)

```
